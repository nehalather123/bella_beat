```python
# Basic information of the dataset
daily_activity.info()
```

```python
# Number of unique ID's
unique_id = len(pd.unique(daily_activity['Id']))
print("Number of unique Id's are " + str(unique_id))
```

##### Observation -
1. There are no null or missing values
2. Dataframe 'daily_activity' has 940 entries and 15 columns
3. Data type of column ActivityDate is wrongly classified as object and needs to be converted to datatime64
4. There are 33 unique Id's 


```python
# Converting "ActivityDate" to datatime64 dtype and yyyy-mm-dd format
daily_activity['ActivityDate'] = pd.to_datetime(daily_activity['ActivityDate'],format='%m/%d/%Y')
daily_activity.info()
daily_activity['ActivityDate'].head()
```

```python
# rearranging columns and adding columns to make visulization easier
df = daily_activity.reindex(columns=['Id', 'ActivityDate', 'DayOfTheWeek', 'TotalSteps', 'TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes', 'TotalExerciseMinutes', 'TotalExerciseHours', 'Calories'])

# renaming columns
df.rename(columns = {"Id":"id", "ActivityDate":"date", "DayOfTheWeek":"day_of_the_week", "TotalSteps":"total_steps", "TotalDistance":"total_dist", "TrackerDistance":"track_dist", "LoggedActivitiesDistance":"logged_dist", "VeryActiveDistance":"very_active_dist", "ModeratelyActiveDistance":"moderate_active_dist", "LightActiveDistance":"light_active_dist", "SedentaryActiveDistance":"sedentary_active_dist", "VeryActiveMinutes":"very_active_mins", "FairlyActiveMinutes":"fairly_active_mins", "LightlyActiveMinutes":"lightly_active_mins", "SedentaryMinutes":"sedentary_mins", "TotalExerciseMinutes":"total_mins","TotalExerciseHours":"total_hours","Calories":"calories"}, inplace = True)
df['day_of_the_week'] = df['date'].dt.day_name()
df['total_mins'] = df['very_active_mins'] + df['fairly_active_mins'] + df['lightly_active_mins'] + df['sedentary_mins']
df['total_hours'] =round(df['total_mins']/60)
df.head()
```

### 5. Share
Creating data visualizations and communicating findings
#### 5.1 Data Visualizations and Findings

```python
#plotting histogram
plt.style.use('default')
plt.figure(figsize=(6,4))
plt.hist(df.day_of_the_week,bins=7,width=0.6,color='blue',edgecolor='black')
plt.xlabel('Day of the week')
plt.ylabel('Frequency')
plt.title('Number of times users logged in application across the week')
plt.grid(True)
plt.show()
```

#### Frequency of usage across the week
In this hitogram, frequency of the usage of the FitBit application can be observed in terms of days of the week.<br>
1. We discovered that users prefer or remember to track their activity on the application during midweek i.e. from Tuesday to Friday.
2. Noting that the frequency dropped towards the weekends and on first day of the week.

```python
# plotting scatter plot
plt.style.use('default')
plt.figure(figsize=(8,6))
plt.scatter(df.total_steps,df.calories,alpha=0.8,c=df.calories,cmap='Spectral')
median_calories = 2304
median_steps = 7638

plt.colorbar(orientation = 'vertical')
plt.axvline(median_steps,color='Blue',label='Median Steps')
plt.axhline(median_calories,color='Red',label='Median calories burned')
plt.xlabel('Steps taken')
plt.ylabel('Calories burned')
plt.title('Calories burned for every step taken')
plt.grid(True)
plt.legend()
plt.show()
```

__Calories burned for every step taken__<br>
From the scatter plot it can be observed that:<br>
1. It is a positive correlation.
2. Intensity of calories burned increases when users are at the range of >0 to 15000 steps with calories burn rate cooling down from 15,000 steps onwards.
3. Noted a few outliers:
    1. Zero steps with zero to minimal calories burned.
    2. 1 observation with more than 35,000 steps but less than 3,000 calories
    3. Deduced that outliers could be due to natural variation of data, change in user's usage or errors in data collection.

```python
# plotting scatter plot
plt.scatter(df.total_hours,df.calories,alpha=0.8,c=df.calories,cmap='Spectral')
median_calories = 2304
median_hours = 20
median_sedentary = 991/60
plt.colorbar(orientation='vertical')
plt.axvline(median_hours,color='Blue',label='Median steps')
plt.axvline(median_sedentary,color='Purple',label='Median sedentary')
plt.axhline(median_calories,color='Red',label='Median hours')
plt.xlabel('Hours logged')
plt.ylabel('Calories burned')
plt.title('Calories burned for every hour logged')
plt.legend()
plt.grid(True)
plt.show()
```

__Calories burned for every hour logged__<br>
The scatter plot show:
1. A weak positive correlation whereby the increase of hours logged does not translate to more calories burned. That is largely due to the average sedentary hours.
2. Few outliers can be observed
    1. zero value outliers
    2. unusual red dot at 24 hours with zero calorie burned.

```python
# calculating total minutes 
very_active_mins = df["very_active_mins"].sum()
fairly_active_mins = df["fairly_active_mins"].sum()
lightly_active_mins = df["lightly_active_mins"].sum()
sedentary_mins = df["sedentary_mins"].sum()
```

```python
# plotting pie chart
slices = [very_active_mins,fairly_active_mins,lightly_active_mins,sedentary_mins]
labels = ['Very active minutes','Fairly active minutes','Lightly active minutes','Sedentary minutes']
colours = ['lightcoral','yellowgreen','lightskyblue','darkorange']
explode = [0,0,0,0.1]
plt.style.use('default')
plt.pie(slices,labels=labels,colors=colours,wedgeprops={'edgecolor':'black'},explode=explode,autopct='%1.1f%%')
plt.title('Percentage of Activity in Minutes')
plt.tight_layout()
plt.show()
```

__Percentage of Activity in Minutes__<br>
As seen from the pie chart,
1. Sedentary minutes take the biggest slice at 81.3%
2. This indicates that users are using the FitBit application to log daily activities such as commute, inactive movements or running errands.
3. Application is rarely being used to track fitness as per the minor percentage of fairly active activities (1.1%) and very active activity (1.7%). This is highly discouraging as FitBit application was developed to encourage fitness.
